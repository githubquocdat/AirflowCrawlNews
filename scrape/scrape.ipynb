```python
list_link_chu_de=[]
def chu_de():
    list_ten_chu_de=[]
    list_ten_chuyen_muc=[]
    driver.get("https://vnexpress.net")

    links = driver.find_elements(By.CSS_SELECTOR,'ul.cat-menu li [href]')

    for link in links:
        if link.get_attribute('href') not in list_no_scrapp  and link.get_attribute('href') not in list_cm:
            list_ten_chu_de.append(link.get_attribute('title'))
            list_link_chu_de.append(link.get_attribute('href'))
        if link.get_attribute('href') in list_cm:
            list_ten_chuyen_muc.append(link.get_attribute('title'))
            # print(link.get_attribute('href'))

    dict_cd = {'ten_Chu_De': list(set(list_ten_chu_de))}
    dict_cm = {'ten_Chuyen_Muc': list_ten_chuyen_muc}

    df_cm = pd.DataFrame(dict_cm)
    df_cd = pd.DataFrame(dict_cd)

    df_cd['id_cd'] = range(1, len(list(set(list_ten_chu_de))) + 1)
    df_cm['id_cm'] = range(1, len(list_ten_chuyen_muc) + 1)

    df_cd=df_cd[['id_cd','ten_Chu_De']]
    df_cm=df_cm[['id_cm','ten_Chuyen_Muc']]


    df_cd.to_csv(f'chu_de_{str(date.today())}.csv',index=False)
    df_cm.to_csv(f'chuyen_muc_{str(date.today())}.csv',index=False)
```

```python
def combine_text(file_path,output_path):
  with open(file_path, 'r', encoding='utf-8') as file:
    text = file.read().strip()  # Read and remove leading/trailing whitespace
    paragraphs = text.split('\n\n') # Split into paragraphs
    combined_lines = [' '.join(paragraph.split()) for paragraph in paragraphs]
    combined_text= '\n'.join(combined_lines)
    with open(output_path, 'w', encoding='utf-8') as output_file:
      output_file.write(combined_text)
```

```python
import csv

def doi_thu_tu_cot(input_file, output_file, new_order):
    """
    Đổi thứ tự các cột trong file CSV theo thứ tự mới.

    Args:
        input_file (str): Đường dẫn đến file CSV đầu vào.
        output_file (str): Đường dẫn đến file CSV đầu ra.
        new_order (list): Danh sách các chỉ số cột theo thứ tự mới.

    Returns:
        None
    """

    with open(input_file, 'r', encoding='utf-8') as infile, open(output_file, 'w', newline='', encoding='utf-8') as outfile:
        reader = csv.reader(infile)
        writer = csv.writer(outfile)

        header = next(reader)  # Đọc dòng tiêu đề
        new_header = [header[i] for i in new_order]  # Sắp xếp lại tiêu đề
        writer.writerow(new_header)  # Ghi dòng tiêu đề mới

        for row in reader:
            new_row = [row[i] for i in new_order]  # Sắp xếp lại dữ liệu từng dòng
            writer.writerow(new_row)  # Ghi dòng dữ liệu mới
```

```python
import csv

def doi_cot_thanh_gia_tri(input_file, output_file, column_name, new_value):
    """
    Đổi tất cả các giá trị trong cột được chỉ định thành giá trị mới trong file CSV.

    Args:
        input_file (str): Đường dẫn đến file CSV đầu vào.
        output_file (str): Đường dẫn đến file CSV đầu ra.
        column_name (str): Tên của cột cần xử lý.
        new_value: Giá trị mới để thay thế.
    """

    with open(input_file, 'r', encoding='utf-8') as infile, open(output_file, 'w', newline='', encoding='utf-8') as outfile:
        reader = csv.DictReader(infile)
        writer = csv.DictWriter(outfile, fieldnames=reader.fieldnames)
        writer.writeheader()

        for row in reader:
            row[column_name] = new_value  # Đặt giá trị thành giá trị mới
            writer.writerow(row)
```

```python
import csv

def drop_columns(input_file, output_file, columns_to_drop):
  """Drops specified columns from a CSV file.

  Args:
    input_file (str): The path to the input CSV file.
    output_file (str): The path to the output CSV file.
    columns_to_drop (list): A list of column names (or indices) to drop.
  """
  with open(input_file, 'r') as infile, open(output_file, 'w', newline='') as outfile:
    reader = csv.reader(infile)
    writer = csv.writer(outfile)

    # Determine column indices to keep
    header = next(reader)  # Read the header row
    indices_to_keep = [i for i, col_name in enumerate(header) if col_name not in columns_to_drop]

    # Write header row (with dropped columns)
    writer.writerow([header[i] for i in indices_to_keep])

    # Write data rows (excluding dropped columns)
    for row in reader:
      writer.writerow([row[i] for i in indices_to_keep])

# Example usage:
input_file = r'D:\emdenthoi\scrape\csv\tac_gia_2024-04-09.csv'
output_file = r'D:\emdenthoi\scrape\csv\tac_gia_2024-04-09_moi.csv'
columns_to_drop = ['id_au']  # Replace with your column names
drop_columns(input_file, output_file, columns_to_drop)
```
